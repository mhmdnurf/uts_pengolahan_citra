```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca gambar asli
image_path = './foto.jpeg'
image = cv2.imread(image_path)

# Pastikan gambar berhasil dibaca
if image is None:
    print("Gambar tidak terdeteksi")
else:
    plt.figure(figsize=(12, 4))

    # Menampilkan gambar asli
    plt.subplot(131)
    plt.title('Gambar Asli')
    # Mengonversi gambar ke grayscale sebelum menampilkannya
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    plt.imshow(grayscale_image, cmap='gray')

    # Menambahkan Gaussian noise
    mean = 0
    std_dev = 25
    noisy_image = grayscale_image + np.random.normal(mean, std_dev, grayscale_image.shape).astype(np.uint8)

    # Menampilkan gambar dengan Gaussian noise
    plt.subplot(132)
    plt.title('Gaussian Noise')
    plt.imshow(noisy_image, cmap='gray')

    # Mengurangi noise dengan filter Gaussian
    kernel_size = (5, 5)
    denoised_image = cv2.GaussianBlur(noisy_image, kernel_size, 0)

    # Menampilkan gambar yang telah direduksi noise
    plt.subplot(133)
    plt.title('Hasil Reduksi Noise (Gaussian Filter)')
    plt.imshow(denoised_image, cmap='gray')

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca gambar dari file
image = cv2.imread('./foto.jpeg')

# Mengubah gambar menjadi grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Membuat salinan gambar grayscale
noisy_image = gray_image.copy()

# Mencari jumlah piksel yang akan diubah menjadi noise
amount_of_noise = 0.1

# Mengacak lokasi piksel yang akan diubah menjadi noise "salt"
num_salt = np.ceil(amount_of_noise * gray_image.size)
salt_coords = [np.random.randint(0, i - 1, int(num_salt)) for i in gray_image.shape]
noisy_image[salt_coords[0], salt_coords[1]] = 255  # Menjadi piksel putih

# Mengacak lokasi piksel yang akan diubah menjadi noise "pepper"
num_pepper = np.ceil(amount_of_noise * gray_image.size)
pepper_coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in gray_image.shape]
noisy_image[pepper_coords[0], pepper_coords[1]] = 0  # Menjadi piksel hitam

# Menggunakan filter median untuk mengurangi noise
filtered_image = cv2.medianBlur(noisy_image, 5)  # Angka 5 adalah ukuran kernel filter

# Menampilkan gambar grayscale asli, gambar dengan noise, dan gambar setelah reduksi noise
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.title('Gambar Asli')
plt.imshow(gray_image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Gambar dengan Noise Salt and Pepper')
plt.imshow(noisy_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Gambar setelah Reduksi Noise')
plt.imshow(filtered_image, cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
image_path = './foto.jpeg'
image = cv2.imread(image_path)

# Pastikan gambar berhasil dibaca
if image is None:
    print("Gambar tidak terdeteksi")
else:
    # Menambahkan kecerahan
    brightness_increase = cv2.convertScaleAbs(image, alpha=1.0, beta=50)
    # Mengurangi kecerahan
    brightness_decrease = cv2.convertScaleAbs(image, alpha=1.0, beta=-50)

    # Menampilkan gambar asli
    plt.figure(figsize=(18, 14))
    plt.subplot(231)
    plt.title('Gambar Asli')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # Menampilkan histogram gambar asli
    plt.subplot(234)
    plt.title('Histogram Gambar Asli')
    for i, color in enumerate(['b', 'g', 'r']):
        hist = cv2.calcHist([image], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)

    # Menampilkan gambar setelah penambahan kecerahan
    plt.subplot(232)
    plt.title('Penambahan Brightness')
    plt.imshow(cv2.cvtColor(brightness_increase, cv2.COLOR_BGR2RGB))

    # Menampilkan histogram gambar setelah penambahan kecerahan
    plt.subplot(235)
    plt.title('Histogram Setelah Kecerahan Ditambahkan')
    for i, color in enumerate(['b', 'g', 'r']):
        hist = cv2.calcHist([brightness_increase], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)

    # Menampilkan gambar setelah pengurangan kecerahan
    plt.subplot(233)
    plt.title('Pengurangan Brightness')
    plt.imshow(cv2.cvtColor(brightness_decrease, cv2.COLOR_BGR2RGB))

    # Menampilkan histogram gambar setelah pengurangan kecerahan
    plt.subplot(236)
    plt.title('Histogram Setelah Kecerahan Dikurangi')
    for i, color in enumerate(['b', 'g', 'r']):
        hist = cv2.calcHist([brightness_decrease], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar berwarna
image_path = './foto.jpeg'
image = cv2.imread(image_path)

# Pastikan gambar berhasil dibaca
if image is None:
    print("Gambar tidak terdeteksi")
else:
    # Mengonversi citra ke citra negatif
    negative_image = cv2.bitwise_not(image)

    # Menampilkan citra negatif
    plt.subplot(121)
    plt.title('Citra Negatif')
    plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))

    # Menampilkan citra asli
    plt.subplot(122)
    plt.title('Gambar Asli')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    plt.show()
```
